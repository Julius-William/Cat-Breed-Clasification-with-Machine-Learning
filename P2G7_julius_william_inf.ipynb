# Inferencing

## Import Package

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.efficientnet import preprocess_input
```

```python
IMG_SIZE = 224
MODEL_PATH = "best_model.keras"   # model yang sudah kamu save

# urutannya harus sama dengan waktu training
label_names = [
        "bengal",
        "domestic_shorthair",
        "maine_coon",
        "ragdoll",
        "siamese"
]
```

```python
model = load_model(MODEL_PATH)
```

```python
def predict_image(image_path):
    print(f"Processing image: {image_path}")

    # Load image
    img = load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))

    # Convert to array
    img_arr = img_to_array(img)

    # EfficientNet preprocessing
    img_arr = preprocess_input(img_arr)

    # Add batch dimension
    img_arr = np.expand_dims(img_arr, axis=0)
    
    # Predict
    pred = model.predict(img_arr)
    pred_class = np.argmax(pred, axis=1)[0]
    pred_label = label_names[pred_class]

    print(f"Prediction: {pred_label}")

    return pred_label
```

## Prediction

Untuk inferencing ini akan dilakukan ke 2 foto kucing yang baru. Berikut adalah foto yang akan diprediksi.

![Domestic_Shorthair](gambar\Dom_Short.jpg)

![Ragdoll](gambar\ragdoll_cat.jpg)

Gambar yang pertama adalah contoh domestic shorthair dan yang kedua adalah contoh ragdoll.

```python
if __name__ == "__main__":

    image_list = [
        "gambar\Dom_Short.jpg","gambar\\ragdoll_cat.jpg"
    ]

    for img_path in image_list:
        predict_image(img_path)
```

Dari hasil inferencingnya dapat dilihat bahwa model dapat dengan bener mengklasifikasi kedua ras kucing yang ada pada gambar tersebut. Menurut saya model ini sudah cukup baik dalam mengklasifikasi ras kucing.
